### My solution

```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1
x = np.linspace(-10, 10, 2000)

y = (0) * (x < 0.6)\
    + (-0.6 + x) * ((x >= 0.6 ) & (x < 4.6))\
    + (8.6 - x) * ((x >= 4.6) & (x < 8.6) )\
    + (0) * (x >= 8.6)

plt.figure()
plt.plot(x, y, label="f(x)", color="red")
plt.title("Problem 1: Piecewise Continuous Function ")
plt.xlabel("x values")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
plt.show()




```

```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 2
x = np.linspace(-10, 10, 2000)

y = (-286.31)*(x < -5.2) \
    + (1.0*x**3 - 4.58*x**2 + 4.18*x - 0.12)*((x >= -5.2) & (x < 6.0)) \
    + (76.08)*(x >= 6.0)

plt.figure()
plt.plot(x, y, label="f(x)", color="red")
plt.title("Problem 2: Piecewise Continuous Function")
plt.xlabel("x values")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
